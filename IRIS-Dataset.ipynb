# IRIS DATASET VISUALIZATION (SEA BORN, MATPLOTLIB)

<p style='text-align:justify;'>The Iris dataset is a popular and widely-used dataset in the field of machine learning and statistics. It is named after the iris flower, as it was initially introduced by the British statistician and biologist Ronald Fisher in 1936 to demonstrate his discriminant analysis technique. The dataset consists of measurements of different features of iris flowers belonging to three different species: Setosa, Versicolor, and Virginica.

The Iris dataset contains 150 samples, with 50 samples for each of the three species. For each sample, there are four features or measurements taken: sepal length, sepal width, petal length, and petal width. These measurements are in centimeters. Based on these features, the goal is to develop a model that can classify an iris flower into one of the three species based on its measurements.

The dataset is often used as a benchmark for classification algorithms and is commonly employed in machine learning tutorials, textbooks, and research papers. It is a relatively small dataset, making it easy to work with, and its features are numeric, which simplifies the process of data analysis and model training.

The Iris dataset is widely recognized for its simplicity, yet it has proven to be valuable for testing and comparing various machine learning algorithms. Its popularity stems from the fact that it serves as a good introductory dataset to learn and understand classification techniques.</p>

## Violin Plot

It is used to visualize the distribution of data and its probability distribution.This chart is a combination of a Box Plot and a Density Plot that is rotated and placed on each side, to show the distribution shape of the data. The thick black bar in the centre represents the interquartile range, the thin black line extended from it represents the 95% confidence intervals, and the white dot is the median.Box Plots are limited in their display of the data, as their visual simplicity tends to hide significant details about how values in the data are distributed

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=iris)
```

# Stacked Histogram

```python
iris['Species'] = iris['Species'].astype('category')
#iris.head()
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```

With Stacked Histogram we can see the distribution of Sepal Length of Different Species together.This shows us the range of Sepal Length for the three different Species of Iris Flower.
